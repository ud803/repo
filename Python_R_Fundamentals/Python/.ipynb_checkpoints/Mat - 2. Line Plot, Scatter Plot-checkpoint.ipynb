# Simple Line Plots

가장 간단한 $ y = f(x) $ 를 그림으로 나타내보자.

먼저 우리가 사용할 함수를 불러오고, 설정을 만지는 것부터 시작하자.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

모든 Mat 도표에서, 먼저 **figure**와 **axes**를 만드는 것부터 시작한다.

```python
fig = plt.figure()
ax = plt.axes()
```

**figure**는 axes, graphics, text, label을 포함하는 하나의 용기라고 볼 수 있다.
**axes**는 위에서 우리가 보고 있는 것들이다 : 라벨이 있는 격자형 상자.

이제 `ax.plot` 함수를 사용해서 데이터를 넣어보자.

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

만약 여러 선을 그리고 싶다면, `plot`함수를 여러번 부르면 된다.

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

### Adjusting the Plot : Colors and Styles

이제 선의 색과 스타일을 바꿔보자.

`plt.plot()` 함수는 이 용도로 추가적인 인자를 받는다.

색을 변경하려면 `color`를 사용한다.

```python
plt.plot(x, np.sin(x - 0), color='blue')        # 이름으로
plt.plot(x, np.sin(x - 1), color='g')           # color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # 0과 1 사이의 그레이스케일
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code 
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB 튜플
plt.plot(x, np.sin(x - 5), color='chartreuse') 
```

스타일을 변경하려면 `linestyle` 인자를 사용한다.

```python
rep = list([0, 1, 2, 3])
rep2 = list(['solid', 'dashed', 'dashdot', 'dotted'])

for num, style in zip(rep, rep2) :
    plt.plot(x, x+ num, linestyle=style)
```

```python
# or equivalently,
rep = list([0, 1, 2, 3])
rep2 = list(['-', '--', '-.', ':'])

for num, style in zip(rep, rep2) :
    plt.plot(x, x+ num, linestyle=style)
```

```python
# Color와 Style을 섞을 수도 있다
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
```

### Adjusting the Plot : Axes Limit

`axes`는 자동으로 영역을 정해주지만, 직접 정할 수도 있다.

`plt.xlim()`과 `plt.ylim()`을 사용한다.

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
# 만약 축을 반대로 하고 싶다면, 반대로 쓰면 된다
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

`plt.axis()` 함수는 한 번에 `xlim`과 `ylim`을 설정할 수 있게 해준다.

`[xmin, xmax, ymin, ymax]`

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

또한 현재 도표가 꽉차도록 자동 설정도 가능하다.

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

혹은 동등한 비율로 보여주는 `equal` 옵션도 있다.

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

### Labeling Plots

도표의 **제목, 축의 제목, 범주**를 넣는 방법을 알아보자.

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
```

추가적인 인자를 통해 라벨의 위치를 조정할 수 있다. (Doc 참고)

Mat에서는 범주를 자동으로 만들어주는 기능이 있다.

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend()
```

MATlAB 스타일과 OO 스타일간의 변형은 아래와 같다.

- `plt.xlabel()` -> `ax.set_xlabel()`
- `plt.xlim()` -> `ax.set_xlim()`
- `plt.title()` -> `ax.set_title()`


```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0,10), ylim=(-2,2),
      xlabel='x', ylabel='sin(x)',
      title='A Simple Plot')
```

### 위는 MATLAB 스타일
### 아래는 Object Oriented 스타일

```python
plt.plot(x, np.sin(x))
plt.xlim(0, 10)
plt.ylim(-2,2)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("A Simple Plot")
```

# Scatter Plot

앞부분에서는 `plt.plot`이나 `ax.plot`을 사용하여 선을 그렸다.

같은 함수를 이용해서 산점도를 나타낼 수 있다.


```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
```

<br>함수에서 사용된 세 번째 인자(`o`)는 사용할 모양을 나타낸다.


```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd'] :
    plt.plot(rng.rand(5), rng.rand(5), marker,
            label="marker='{0}'".format(marker))
```

<br> 추가적으로, 이 코드들은 선과 함께 사용될 수 있다!


```python
# line('-'), circle ('o'), black('k')
plt.plot(x, y, '-ok')
```

`plt.plot`과 더불어 쓰이는 방법은 `plt.scatter`도 있다.

```python
plt.scatter(x, y, marker='o')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```
