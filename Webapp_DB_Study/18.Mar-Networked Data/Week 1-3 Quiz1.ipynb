### Q3. 아래 XML 데이터를 읽어와라

zipcode가 21231인 레스토랑은 몇 개나 있는가?

```r
library(XML)
```

```r
doc <- xmlTreeParse("http://d396qusza40orc.cloudfront.net/getdata%2Fdata%2Frestaurants.xml",
                  useInternal=TRUE)
rootNode <- xmlRoot(doc)
```

```r
rootNode
```

```r
ans = xpathSApply(rootNode, "//zipcode", xmlValue)
```

```r
table(ans == "21231")
```
