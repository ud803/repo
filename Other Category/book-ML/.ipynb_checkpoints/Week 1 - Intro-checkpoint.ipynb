# 머신 러닝 (Machine Learning)

** 머신 러닝은 별다른 프로그래밍 없이 컴퓨터로 하여금 배우도록 하게 만드는 학문**

Machine learning is the science of getting computers to learn, without being explicitly programmed. 


### 1. 머신러닝의 예
- Database Mining
    - Web click data, medical records, biology, engineering
- Applications can't program by hand
    - Autonomous helicopter
    - Handwriting Recognition
    - NLP
    - Computer Vision
- Self-customizing programs
    - Amazon, Netflix product recommendations
- Understanding human learning


### 2. 머신러닝의 정의

- Arthur Samuel (1959) 
<br>: ***Field of study that gives computers the ability to learn without being explicitly programmed.***
<br>조금더 고전적인 정의이다.


- Tom Mitchell (1998) : 
<br>더 현대적인 정의
<br>    Well-posed Learning Problem 
<br>: ***A computer program is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P, improves with experience E.***
    

    Ex) 이메일이 스팸 작업을 처리할 때, 여기서 'T'에 해당하는 업무는 무엇일까?
    -> 이메일을 스팸 혹은 정상으로 분류하는 업무

    Ex) 체커 게임을 한다고 가정하자. 이 때,
    E = 체커 게임을 반복해서 얻는 경험
    T = 체커 게임을 하는 업무
    P = 다음 게임에서 프로그램이 이길 확률



### 3. 머신러닝 알고리즘
- 지도 학습 (Supervised Learning)
<br> '지도'라는 말을 쓰는 이유는, 우리가 알고리즘에게 **Right Answer**를 제공하기 때문이다.

    - 회귀 분석(Regression): 연속형 결과 값을 예측한다
    - 분류(Classification) : 이산형 결과 값(범주형 결과 값)을 예측한다
<br><br>
- 비지도 학습 (Unsupervised Learning): <br> 주어진 데이터로부터 어떠한 구조나 패턴을 발견할 수 있는지 알고리즘에게 묻는 것이다.

    - 군집 분석(Clustering) : 성질이 비슷한 것들끼리 함께 묶는다
        - 구글에서 뉴스를 같은 토픽별로 묶는 것
        - 소셜네트워크 분석
        - 시장 분할 (Market Segmentation)
        - 천문 데이터 분석
    - Cocktail party problem : 2명이 칵테일 파티에서 대화를 한다. <br>마이크가 두 대 있고, 여러 사람이 말할 때 이들의 목소리를 분리할 수 있을까? 
        - [Cocktail Party Effect](https://en.wikipedia.org/wiki/Cocktail_party_effect)
        - 오디오 레코딩을 듣고 두 레코드를 분리하는 알고리즘
        - 알고리즘 자체는 그저 한 줄로 실행할 수 있다
        

- 기타
    - 강화 학습 (Reinforcement Learning)
    - 추천 시스템 (Recommender System)
