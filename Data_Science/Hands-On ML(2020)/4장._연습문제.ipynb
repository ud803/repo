### Q1. 수백만 개의 특성을 가진 훈련 세트에서는 어떤 선형 회귀 알고리즘을 사용할 수 있을까?

배치, 확률적, 미니배치 경사 하강법을 사용할 수 있다. 일반적인 정규방정식의 경우, 특성의 수가 커지면 해를 구하는 데 시간이 오래걸리지만,

경사 하강법을 이용하는 경우는 특성의 수에 영향을 적게 받는다.


### Q2. 훈련 세트에 있는 특성들이 각기 아주 다른 스케일을 가지고 있다. 이런 데이터에 잘 작동하지 않는 알고리즘은 무엇일까? 그 이유는? 이 문제를 어떻게 해결할 수 있을까?

기존 내 생각:

정규방정식의 경우, 비용 함수에서 값들을 계산할 때 스케일이 큰 값들의 영향이 더 커져 문제가 될거라고 생각함.


### Q3. 경사 하강법으로 로지스틱 회귀 모델을 훈련시킬 때 지역 최솟값에 갇힐 가능성이 있을까?

없다. 로지스틱 회귀 모델의 비용 함수는 볼록 함수이기 때문에 전역 최솟값만 존재한다.


### Q4. 충분히 오랫동안 실행하면 모든 경사 하강법 알고리즘이 같은 모델을 만들어낼까?

그렇다고 본다.

### Q5. 배치 경사 하강법을 사용하고 에포크마다 검증 오차를 그래프로 나타내보았다. 검증 오차가 일정하게 상승되고 있다면 어떤 일이 일어나고 있는 걸까? 이 문제를 어떻게 해결할 수 있을까?

과소적합의 경우로써, 데이터에 잘 적합되지 않은 경우이다. 데이터의 양을 늘리거나, 모델의 파라미터가 더 많은 모델을 쓰거나, 규제를 줄인다.

### Q6. 검증 오차가 상승하면 미니배치 경사 하강법을 즉시 중단하는 게 가장 좋은 방법일까?

미니배치의 경우, 데이터의 일부를 살피기 때문에 더 지켜봐야 한다.


### Q7. 어떤 경사 하강법 알고리즘이 가장 빠르게 최적 솔루션의 주변에 도달할까? 실제로 수렴하는 것은? 다른 방법들도 수렴하게 만들 수 있을까?

배치 경사 하강법이 가장 빠르게 도달할 것이다. 실제 수렴은 SGD가 더 빠르다. 학습률을 점진적으로 조정하거나, 조기 종료를 통해 할 수 있다.

### Q8. 다항 회귀를 사용했을 때 학습 곡선을 보니 훈련 오차와 검증 오차의 간격이 크다. 무슨 일이 생긴걸까? 이 문제를 해결하는 세 가지 방법은?

과대적합된 경우이다. 데이터의 양을 늘리거나, 규제를 늘리거나, 단순한 모델을 사용한다.

### Q9. 릿지 회귀를 사용했을 때, 훈련 오차와 검증 오차가 둘 다 비슷하고 높았다. 이 모델에는 높은 편향이 문제일까, 아니면 높은 분산이 문제일까? 규제 하이퍼파라미터 알파를 증가시켜야 할까, 아니면 줄여야 할까?

과소적합된 경우에 가깝다. 이는 모델이 데이터에 잘 학습되지 않았기에, 높은 편향이 문제가 된다. 하이퍼파라미터 알파를 감소시켜 모델을 더 자유롭게 풀어줘야 한다.

### Q11. 사진을 낮과 밤, 실내와 실외로 분류하려고 한다. 두 개의 로지스틱 회귀 분류기를 만들어야 할까, 아니면 하나의 소프트맥스 회귀 분류기를 만들어야 할까?

낮/밤과 실내/실외는 모두 서로다른 케이스이다. 이는 별개의 모델이고, 사용되는 데이터(구체적으로는, 사용되는 특성)도 다를 것이기 때문에 두 개의 로지스틱 회귀 분류기를 만든다.
