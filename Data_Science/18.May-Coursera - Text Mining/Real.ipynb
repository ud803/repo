```python
import re
import pandas as pd


doc = []
with open('dates.txt') as file:
    for line in file:
        doc.append(line)

def date_sorter():
    

    p = re.compile(r'\
(?P<Month>[01]?[0-9]|Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[- ,.:/]+\
(?P<Date>[0123]?[0-9])[a-z]*[- ,.:/]+\
(?P<Year>[12]?[09]?[0-9]{2})\
')

    pp = re.compile(r'\
(?P<Date>[0123]?[0-9])[a-z]*[- ,.:/]+\
(?P<Montg>[01]?[0-9]|Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[- ,.:/]+\
(?P<Year>[12]?[09]?[0-9]{2})\
')

    q = re.compile(r'\
(?P<Month>[01]?[0-9]|Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[- ,.:/]+\
(?P<Year>[12]?[09]?[0-9]{2})\
')

    r = re.compile(r'\
(?P<Year>[12]?[09]?[0-9]{2})\
')

    temp = []
    ptemp = []
    pptemp=[]
    qtemp = []
    rtemp = []
    for i in range(500) :
        ptemp.append(-1)
        qtemp.append(-1)
        rtemp.append(-1)
        pptemp.append(-1)
    for i in range(500) :
        if(p.search(doc[i])) :
            ptemp[i] = p.findall(doc[i])
    for i in range(500) :
        if(pp.search(doc[i])) :
            pptemp[i] = pp.findall(doc[i])

    for i in range(500) :
        if(q.search(doc[i])) :
            qtemp[i] = q.findall(doc[i])
    for i in range(500) :
        if(r.search(doc[i])) :
            rtemp[i] = r.findall(doc[i])

    indexl = []
    for i in range(500) :
        indexl.append(0)

    for i in range(500) :
        if(ptemp[i] != -1) :
            indexl[i] = ptemp[i]
    for i in range(500) :
        if(pptemp[i] != -1) :
            indexl[i] = pptemp[i]        
    for i in range(500) :
        if(qtemp[i] != -1) :
            if(indexl[i]==0):
                indexl[i] = qtemp[i]
    for i in range(500) :
        if(rtemp[i] != -1) :
            if(indexl[i]==0):
                indexl[i] = rtemp[i]

    real = []

    for item in indexl :
        if(len(item[0]) !=4) :
            real.append(list(item[0]))
        else :
            real.append(item)

    for i in range(500):
        real[i].append("s{0}".format(i))

    indexl


    num1 = []
    num2 = []
    num3 = []

    for item in real :
        if (len(item) == 2) :
            num1.append(item)
        elif (len(item) ==3) :
            num2.append(item)
        elif (len(item) ==4) :
            num3.append(item)

    for item in num1 :
        temp = item[0]
        item.append('1')
        item.append('1')
        item[0], item[2] = item[2], item[0]
        item[1], item[3] = item[3], item[1]


    dictt = {'Jan' :'1', 'Feb' :'2', 'Mar':'3', 'Apr':'4', 'May':'5', 'Jun':'6', 'Jul':'7',\
            'Aug':'8', 'Sep':'9', 'Oct':'10', 'Nov':'11', 'Dec':'12'}

    for i in range(len(num2)) :
        if(num2[i][0].isalpha()) :
            num2[i][0] = dictt[num2[i][0]]
    for i in range(len(num2)) :
        if(num2[i][0].startswith('0')) :
            num2[i][0] = num2[i][0][1:]
        num2[i].insert(1,'1')

    for i in range(len(num3)) :
        if(num3[i][0].isalpha()) :
            num3[i][0] = dictt[num3[i][0]]
    for i in range(len(num3)) :
        if(len(num3[i][2]) !=4) :
            num3[i][2] = '19' + num3[i][2]
    for i in range(len(num3)) :
        if(num3[i][1].isalpha()) :
            num3[i][1] = dictt[num3[i][1]]
            num3[i][0], num3[i][1] = num3[i][1], num3[i][0]
    for i in range(len(num3)) :
        if(num3[i][0].startswith('0')) :
            num3[i][0] = num3[i][0][1:]
        if(num3[i][1].startswith('0')) :
            num3[i][1] = num3[i][1][1:]


    real = sorted(real, key= lambda x: (x[2], x[0], x[1]))

    for i in range(500) :
        real[i].append(i)

    answer = []
    for i in range(500) :
        answer.append(0)

    for i in range(500) :
        answer[int(real[i][3][1:])] = real[i][4]

    answer = pd.Series(answer)
    return (answer)
```
